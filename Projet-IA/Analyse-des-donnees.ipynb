# Nettoyage et filtrage des donnees

```python

# Créer une nouvelle fonction qui détermine si la valeur en paramètre est manquante:
def num_missing(x):
    return sum(x.isnull())

# On applique cette fonction pour chaque colonne:
print("Valeurs manquantes par colonne:")
print(ds_df.apply(num_missing, axis=0)) #axis=0 définit que la fonction sera bien appliquée sur chaque colonne

# Puis application pour chaque ligne:
print("\nValeurs manquantes par ligne:")
print(ds_df.apply(num_missing, axis=1).head()) #axis=1 définit que la fonction sera bien appliquée sur chaque ligne
```

```python
#afficher les valeurs 
s = ds_df.isna().sum()
s = s[s != 0].sort_values(ascending=False)
s
```

```python
#calculer le pourcentage des valeurs nulles de chaque colonne
percent_missing=ds_df.isna().sum()/len(ds_df)*100
```

```python
#Afficer les colonnes qui ont un pourcentage  des valeurs nulls >= 10% 
ax=percent_missing[percent_missing>=10].sort_values(ascending=False).plot.bar(title='Pourcentage des valuesNaN ',figsize=(5,5));
ax.set_ylabel('% of NaN elements');
```

```python
#Cette fonction prend tous les espaces entre les mots des noms de colonnes et les remplace avec '_' .
#L'en-tête de l'argument correspond à un nom de colonne. 

def remove_space(header):

    list1=list()
    words_header=header.split()    
    size=int(len(words_header))
    
    for i in range(len(words_header)):
        if i<size-1:
            list1.append(words_header[i]+'_')            
        else:
            list1.append(words_header[i])
                
    separator = ''
    final=separator.join(list1)    
    
    return final
```

```python
#Cette fonction prend tous les caractères spéciaux présents dans les noms de colonnes et les remplace
#avec un autre caractère selon le cas.
#L'en-tête de l'argument correspond à un nom de colonne.

def remove_sp_char(header):

    if "'" in header:
        header=header.replace("'",'')
    if "," in header:
        header=header.replace(",",'.')
    if "_-_" in header:
        header=header.replace("_-_",'_')
    if "/" in header:
        header=header.replace("/",'_or_')
    if ":" in header:
        header=header.replace(":",'')
    if "-" in header:
        header=header.replace("-",'_')
        
    return header
```

```python
list_new_head=list()
headers=ds_df_record.columns

for header in headers:
    header1=remove_space(header) # Les espaces sont remplacée
    header1=header1.casefold()   # Toutes les lettres Majus sont changées
    header1=remove_sp_char(header1) # Les caracteres speciaux characters sont remplacée
        
    list_new_head.append(header1)
```
